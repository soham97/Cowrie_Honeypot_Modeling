```python
import pandas as pd
import numpy as np
import hmm
from sklearn.preprocessing import LabelEncoder
from tqdm import tqdm_notebook as tqdm
import pickle
```

```python
with open('seq_events.txt','rb') as file:
    k = pickle.load(file)
```

```python
data = []
```

```python
for i in tqdm(range(0,len(k))):
    if len(k[i]) == 4:
        data.append(k[i])
        
len(data)
```

```python
data = np.array(data)
data.shape
```

```python
data
```

```python
N = 15
M = 15
T = 4
```

```python
mod = hmm.HMM(N, M, T)
```

```python
print (mod.train(data))
```

```python
prob = []
for i in tqdm(range(0,len(data))):
    prob.append(mod.full_prob(mod.forward(data[i])))
```

```python
df = pd.DataFrame()
df['prob'] = prob
df.describe()
```
